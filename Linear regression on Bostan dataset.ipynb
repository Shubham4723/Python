```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error ,r2_score
```

```python
X_Train, X_Test, Y_Train, Y_Test= train_test_split(X_feat, Y, test_size=.2, random_state=2) 
```

```python
re = LinearRegression()
re.fit(X_Train, Y_Train)
print(re.coef_)
print(re.intercept_)
```

```python
Y_TrainPred=re.predict(X_Train)
print (np.sqrt(mean_squared_error(Y_Train, Y_TrainPred)))  ## RMSE

Y_Pred=re.predict(X_Test)
np.sqrt(mean_squared_error(Y_Test, Y_Pred))
```
